```python
"""
College Event Feedback Analysis - Internship Project (Task 3)

Intern: YASH SOMPURA
Internship Program: Data Science & Analytics
Company: Future Interns

"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('student_feedback.csv')
print("CSV file loaded successfully.")

df.columns = df.columns.str.replace(' ', '_')
print("\nColumn names standardized.")

print("\nDataFrame Head after cleaning:")
print(df.head())

print("\nDataFrame Info:")
df.info()
print("\nMissing Values (should be 0 for all after inspection):")
print(df.isnull().sum())

rating_columns = df.drop(columns=['Student_ID']).columns

average_ratings = df[rating_columns].mean().sort_values(ascending=False)

print("\nAverage Ratings for each feedback aspect (sorted highest to lowest):")
print(average_ratings)

highest_rated_aspect = average_ratings.idxmax()
lowest_rated_aspect = average_ratings.idxmin()

print(f"\nHighest rated aspect: '{highest_rated_aspect}' with an average rating of {average_ratings.max():.2f}")
print(f"Lowest rated aspect: '{lowest_rated_aspect}' with an average rating of {average_ratings.min():.2f}")

#visualization using matplotlib
plt.figure(figsize=(12, 8))
sns.barplot(x=average_ratings.values, y=average_ratings.index, palette='viridis')
plt.title('Average Ratings for Course Feedback Aspects', fontsize=16)
plt.xlabel('Average Rating (Scale 1-10)', fontsize=12) # Assuming a 1-10 scale based on data values
plt.ylabel('Feedback Aspect', fontsize=12)
plt.xlim(0, 10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
